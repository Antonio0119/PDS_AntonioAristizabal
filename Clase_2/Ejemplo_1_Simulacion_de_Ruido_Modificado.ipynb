# Ejemplo 1. Como simular ruido blanco a una señal analógica.

```python
# Importar las librerias necesarias
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 20)     # Vector de tiempo
senal = np.sin(t)                  # Señal
noise = np.random.randn(20)         # ruido blanco
signal_with_noise = senal + noise  # Se añade el ruido a la señal
```

```python
# Grafica de las señales con y sin ruido
plt.figure()
plt.plot(t, senal, 'b', label = 'Señal son ruido')
plt.plot(t, signal_with_noise, 'r', label = 'Señal con ruido')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```

**Nuevo ejemplo de función con ruido:**

```python
t = np.linspace(0, 4*np.pi, 50)     # Vector de tiempo
senal = 5*np.cos(t)                  # Señal
noise = np.random.randn(50)        # ruido blanco
signal_with_noise = senal + noise  # Se añade el ruido a la señal
```

```python
# Grafica de las señales con y sin ruido
plt.figure()
plt.plot(t, senal, 'g', label = 'Señal sin ruido')
plt.plot(t, signal_with_noise, 'r', label = 'Señal con ruido')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.title('Señal analógica sin y con ruido blanco')
plt.show()
```

**Nuevo ejemplo de función señal cuadrada con ruido:**

```python
t = np.linspace(0, 1, 500, endpoint=True)
senal = signal.square(2*np.pi*3*t)

noise = 0.1*np.random.randn(500)        # ruido blanco
signal_with_noise = senal + noise  # Se añade el ruido a la señal

plt.figure()

plt.plot(t, signal_with_noise, 'r', label = 'Señal con ruido', linewidth='1')
plt.plot(t, senal, 'g', label = 'Señal sin ruido', linewidth='1')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.title('Señal analógica sin y con ruido blanco')
plt.show()
```
