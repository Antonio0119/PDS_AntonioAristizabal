```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 4*np.pi, 50)     # Vector de tiempo
signal = 5*np.cos(t)                  # Señal
noise = np.random.randn(50)        # ruido blanco
signal_with_noise = signal + noise  # Se añade el ruido a la señal
```

```python
# Grafica de las señales con y sin ruido
plt.figure()
plt.plot(t, signal, 'g', label = 'Señal sin ruido')
plt.plot(t, signal_with_noise, 'r', label = 'Señal con ruido')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.title('Señal analógica sin y con ruido blanco')
plt.show()
```
