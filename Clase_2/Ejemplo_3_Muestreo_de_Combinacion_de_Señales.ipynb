```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = 50     # Frecuencia de la señal 1
f2 = 150    # Frecuencia de la señal 2
f3 = 200     # Frecuencia de la señal 3
fs = 1e3    # Frecuencia de muestreo
t = np.arange(0, 0.1, 1/fs) # Vector de tiempo
signal = (2*np.cos(2*np.pi*f1*t)) - (30*np.sin(2*np.pi*f2*t)) - (15*np.cos(2*np.pi*f3*t)) # Suma de señales
```

```python
# Grafica de la señal analógica
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal analógica con tasa de Nyquist mínima')
plt.grid()
plt.show()
```

La señal anterior esta muestreada correctamente porque se usó una frecuencia de muestro superior a 2 veces la máxima que es 200Hz, o sea, mayor a 400Hz. Si se usa por ejemplo una frecuencia de muestreo de 100Hz, que es inferior a 400Hz que es la tasa de Nyquist mínima en este caso, la señal estaría mal muestreada como se muestra a continuación.

```python
fs = 1e2    # Frecuencia de muestreo nueva
t = np.arange(0, 0.1, 1/fs) # Vector de tiempo
signal = (2*np.cos(2*np.pi*f1*t)) - (30*np.sin(2*np.pi*f2*t)) - (15*np.cos(2*np.pi*f3*t)) # Suma de señales
```

```python
# Grafica de la señal analógica
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal analógica sin tasa de Nyquist mínima')
plt.grid()
plt.show()
```
