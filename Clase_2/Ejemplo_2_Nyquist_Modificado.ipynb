# Ejemplo 2. Criterio de Nyquist para la frecuencia de muestreo.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

La frecuencia de muestreo debe cumplir el criterio de Nyquist: $F_{s}\geq 2F_{max}$<br>
Suponga una señal cosenoidal con amplitud 5 y frecuencia máxima de 2Hz.

```python
fmax = 2                            # Frecuencia maxima
w = 2*np.pi*fmax                    # Frecuencia en radianes
fs = 2.1*fmax                       # Frecuencia de muestro
t = np.arange(0, 2*np.pi, 1/fs)     # Vector de tiempo
signal = 5*np.cos(w*t)                # Señal senoidal
```

```python
# Grafica de la señal analógica
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal analógica 5cos(wt)')
plt.grid()
plt.show()
```

Aunque la señal no se muestra como se desea, se puede reconstruir la señal ya que cumple el criterio de Nyquist. Veamos que pasa si se aumenta la frecuencia de muestreo a 10 veces la frecuencia máxima.

```python
fs = 10*fmax                        # Frecuencia de muestro aumentada
t = np.arange(0, 2*np.pi, 1/fs)     # nuevo vector de tiempo
signal = 5*np.cos(w*t)                # Nueva señal senoidal
```

```python
# Grafica de la con fs=10*fmax
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal analógica 5cos(wt)')
plt.grid()
plt.show()
```

Ya se puede notar la forma cosenoidal correctamente, si se aumenta mas la frecuencia de muestreo la grafica se va a observar mejor. Para no elevar demasiado el coste computacional es recomendado no elevar mucho este valor.

Veamos que pasa si muestreamos por debajo de la taza de Nyquist:

```python
fs = fmax                           # Frecuencia de muestro aumentada
t = np.arange(0, 2*np.pi, 1/fs)     # nuevo vector de tiempo
signal = 5*np.cos(w*t)                # Nueva señal senoidal
```

```python
# Grafica de la con fs=fmax
plt.figure()
plt.plot(t, signal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal analógica 5cos(wt)')
plt.grid()
plt.show()
```

Ya no hay forma cosenoidal, por ende, no es posible recuperar la señal posteriormente.
