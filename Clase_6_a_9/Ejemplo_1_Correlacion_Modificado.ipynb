# Ejemplo 1. Correlación.


```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 4, 2, 5, 1])
y = np.array([4, 2, 5])
Rxy  = np.correlate(x, y, mode='full')
```

```python
# Grafica de la correlación
plt.figure()
plt.stem(Rxy)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Otro ejemplo de correlación cruzada:

```python
x = np.array([2, -1, 3, 7, 1, 2, -3])
y = np.array([1, -1, 2, -2, 4, 1])
Rxy  = np.correlate(x, y, mode='full')
```
