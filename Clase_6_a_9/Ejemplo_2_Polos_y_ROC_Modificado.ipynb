# Ejemplo 2. Polos y ROC

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sp
```

```python
def zeropoles(b, a=1):
    # Genera una grafica con los polos y ceros de la funci칩n de transferencia
    # y la circunferencia unitaria.
    w,h = sp.freqz(b, a)
    sys1 = sp.lti(b, a)
    ang = np.arange(0.0, 2*np.pi, 0.01)
    xp = np.cos(ang)
    yp = np.sin(ang)
    plt.figure()
    plt.plot(xp, yp, '--')
    plt.plot(sys1.zeros.real, sys1.zeros.imag, 'o')
    plt.plot(sys1.poles.real, sys1.poles.imag, 'x')
    plt.grid()
    plt.show()
```

#### $X(Z)=2 \frac{1-0.5Z^{-1}}{1+2Z^{-2}}=\frac{2Z^{2}-Z}{Z^{2}+2}$

```python
b = [2, -1 , 0] # Numerados
a = [1, 0, 2]   # Denominador
zeropoles(b, a)
```

$Den=1+2Z^{-2} \Rightarrow \frac{-2}{Z^2}<1 \Rightarrow |Z|>\pm j \sqrt{2}$ <br>
La anterior funci칩n es causal-inestable, ya que la regi칩n de convergencia se extiende del polo mas lejano hacia el infinito, esto se ve despues de hacer el an치lisis te칩rico, tambien es inestable porque los polos no estan dentro de la circunferencia unidad, por lo cual la ROC no contiene a la circunferencia unidad.

**Otro ejemplo:** <br>
$H(Z)=\frac{Z^{-1}}{(1-0.5Z^{-1})(1+0.75Z^{-1})} \Rightarrow |Z|>0.5, |Z|<0.75$

```python
b = 1 # Numerados
a = [1, 1/4, -3/8]   # Denominador
zeropoles(b, a)
```

En este caso se puede ver que es No causal-inestable, porque la ROC no se extiende hacia el infinito, adem치s, la ROC no contiene a la circunferencia unidad.

**Nuevo ejemplo:**<br>
$H(Z)=\frac{Z}{(Z-0.5)(Z-0.25)} \Rightarrow |Z|>0.5, |Z|>0.25$

```python
b = [1, 0] # Numerados
a = [1, -3/4, 1/8]   # Denominador
zeropoles(b, a)
```

En este caso se obtiene que es Causal-Estable, puesto que la ROC se extiende hacia el infinito y adem치s contiene la circunferencia unitaria.
