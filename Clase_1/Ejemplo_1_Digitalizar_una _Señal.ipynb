```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 30) # Vector de tiempo
signal = 10*np.cos(30*t)+5      # Señal analogica
```

```python
# Grafica de la señal analógica y digital
plt.figure()
plt.plot(t, signal, 'b', label = 'Señal analógica 10cos(30t)+5')
plt.stem(t, signal, 'g', label = 'Señal digitalizada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.title('Señal Analógica Digitalizada')
plt.show()
```
