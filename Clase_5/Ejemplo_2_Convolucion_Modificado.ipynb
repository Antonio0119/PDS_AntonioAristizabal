# Ejemplo 2. Convolución de Señales.


```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

$x(n)=[\vec{15}, 10, 2]$ <br>
$h(n)=[4, \vec{13}, 7]$

```python
# Sabiendo que h(n) tiene el cero en el 20 y x(n) lo tiene en el primer elemento:
# N1=0 y N2=-1. M1=2, M2=1. N1+N2 <= n <= M1+M2
# El rango debe ser desde [-1,3]
n = np.arange(-1,4)
h = np.array([4, 13, 7])  # h(n)
x = np.array([15, 10, 2])     # x(n)
y = np.convolve(h, x)       # Convolución de las señales y(n)
```

```python
# Grafica de la convolución
plt.figure()
plt.stem(n, y)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Otro ejemplo de convolución:

$j(n)=[\vec{-2}, 0, 1, -1, 3]$ <br>
$k(n)=[\vec{1}, 2, 0, -1]$

```python
# En este caso los ceros de ambos arreglos están en el primer elemento
# entonces se sabe que N1=N2=0 y M1=4, M2=3
# El rango es entonces: 0 <= n <= 7
n2 =np.arange(0, 8)
j = np.array([-2, 0, 1, -1, 3])
k = np.array([1, 2, 0, -1])
z = np.convolve(j, k)
```

```python
# Grafica de la convolución
plt.figure()
plt.stem(n2, z)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Otro ejemplo de convolución:<br>
$k(n)=[1,\vec{2}, 1, 1, 3, 4]$ <br>
$j(n)=[1, \vec{2}, 3]$

```python
# Para este caso ambos ceros están en el 2
# entonces se sabe que N1=-1, N2=-1 y M1=2, M2=4
# El rango es entonces: 0 <= n <= 7
n3 =np.arange(-2, 6)
j = np.array([1, 2, 3])
k = np.array([1, 2, 1, 1, 3, 4])
w = np.convolve(j, k)
```

```python
# Grafica de la convolución
plt.figure()
plt.stem(n3, w)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```
