# Ejemplo 2. Generar señales útiles como el delta, escalón, y rampa.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

Es muy importante tener bien definido el vector de muestras, este se genera en base a N que es el numero de muestras, se recomienda siempre un N impar.

```python
N = 9
n = np.arange(-N//2+1, N//2+1)
```

#### 1. Función delta


```python
delta = np.zeros(N) # Vector de ceros
delta[N//2] = 1     # Se pone un uno en la mitad

# Grafica de la secuencia
plt.figure()
plt.title('Función delta')
plt.stem(n, delta)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

#### 2. Función escalón unitario

```python
amplitud_escalon = 1    # Amplitud
desplazamiento = 1      # Desplazamiento del escalón
u = np.heaviside(n-desplazamiento, amplitud_escalon)    # Función escalón

# Otra forma de hacerlo:
# u = np.piecewise(n, n>=1, [1, 0])

# Grafica de la secuencia
plt.figure()
plt.figure()
plt.title('Función escalón unitario')
plt.stem(n, u)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

#### 3. Función rampa

```python
rampa = []              # Vector vacío para rellenar
desplazamiento2 = 0     # Desplazamiento de la rampa
cont = 1                # Contador para generar la rampa
for i in range(N):
    if (i >= (N//2+1)-(-desplazamiento2)):
        rampa.append(cont)
        cont += 1
    else:
        rampa.append(0)

# Grafica de la secuencia
plt.figure()
plt.figure()
plt.title('Función rampa')
plt.stem(n, rampa)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

#### 4. Función exponencial

```python
N = 7
n = np.arange(0, N)

exponencial= np.exp(-n)
# Grafica de la secuencia
plt.figure()
plt.figure()
plt.title('Función exponencial')
plt.stem(n, exponencial)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```
